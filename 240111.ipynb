```python
# matplotlib으로 그래프, 축 제목 정하기
x = ['A','B','C']
y = [100, 150, 120]
plt.bar(x,y)
plt.title('Test Graph')
plt.xlabel('Group')
plt.ylabel('Score')
plt.show()
```

```python
# matplotlib subplots
x=[1,2,3,4,5]
y=[1,4,9,16,25]

fig, axs = plt.subplots(2,2)
axs[0,0].plot(x,y)
axs[0,1].scatter(x,y)
axs[1,0].bar(x,y)
axs[1,1].barh(x,y)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

#iris = sns.load_dataset("iris")
df1 = pd.read_csv('iris.csv')
df1
```

```python
df1.info()
```

```python
# scatterplot
sns.scatterplot(x='petal_length', y='petal_width',hue='species',data=df1)
```

```python
# countplot
sns.countplot(x='species',data=df1)
plt.show()
```

```python
# barplot
sns.barplot(x='species',y='petal_length',data=df1,errorbar='sd')
```

```python
# pairplot
sns.pairplot(df1,hue='species')
```

```python
# regplot
sns.regplot(x='petal_length',y='petal_width',data=df1)
```

```python
# heatmap
sns.heatmap(data=df1,annot=True,fmt='.2f',cbar=False)
```
